# Data prepare

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
%load_ext nb_black
```

```python
scaler = StandardScaler()
```

# Prepare

```python
train_df = pd.read_csv(
    "../../data/santander-customer-transaction-prediction/train.csv"
).set_index("ID_code")
train_df
```

```python
test_df = pd.read_csv(
    "../../data/santander-customer-transaction-prediction/test.csv"
).set_index("ID_code")
test_df
```

```python
def do_work(df: pd.DataFrame) -> pd.DataFrame:
    df1 = df.copy()

    var_columns = df1.columns
    if "target" in df1.columns:
        var_columns = df1.drop("target", axis=1).columns

    df1[var_columns] = scaler.fit_transform(df1[var_columns])
    return df1
```

```python
train1_df = do_work(train_df)
train1_df
```

```python
train1_df.to_csv("../../data/santander-customer-transaction-prediction/train_1.csv")
```

```python
test1_df = do_work(test_df)
test1_df
```

```python
test1_df.to_csv("../../data/santander-customer-transaction-prediction/test_1.csv")
```
